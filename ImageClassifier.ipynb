```python
import matplotlib.pyplot as plt
import numpy as np
import PIL
import tensorflow as tf
import pandas as pd
import os
import cv2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
batch_size = 32
img_height = 180
img_width = 180

data_dir = 'C:/Users/rgkri/OneDrive/Geetha/GUVI/project/ImageClassifier/images/dataset/training_set' 
test_path = 'C:/Users/rgkri/OneDrive/Geetha/GUVI/project/ImageClassifier/images/dataset/test_set'
```

```python
x_train=[]

for folder in os.listdir(data_dir):

    sub_path=data_dir+"/"+folder

    for img in os.listdir(sub_path):

        image_path=sub_path+"/"+img

        img_arr=cv2.imread(image_path)

        img_arr=cv2.resize(img_arr,(224,224))

        x_train.append(img_arr)
        
        
x_train
```

```python
x_test=[]

for folder in os.listdir(test_path):

    sub_path=test_path+"/"+folder

    for img in os.listdir(sub_path):

        image_path=sub_path+"/"+img

        img_arr=cv2.imread(image_path)

        img_arr=cv2.resize(img_arr,(224,224))

        x_test.append(img_arr)
        
x_test
```

```python
train_x=np.array(x_train)
test_x=np.array(x_test)
#
```
